```python
#cleaning up methods
df['method'] = df['method'].replace(r'"', value="", regex=True)
df['method'] = df['method'].str.strip('"')

methods = pd.Series(df['method'].replace(r"\s/.*", value="", regex=True))

methods = methods.str.strip(".")
methods = methods.str.strip()
methods = methods.str.title()
methods = methods.replace(r"(?i)Trangle.*", value="Triangle", regex=True) #triangle misspelling
methods = methods.replace(r"(?i)Triangulo.*", value="Triangle", regex=True) #triangle misspelling
methods = methods.replace(r"(?i)Trangle.*", value="Triangle", regex=True) #triangle misspelling
methods = methods.replace(r"(?i)Trianglle.*", value="Triangle", regex=True) #triangle misspelling
methods = methods.replace(r"(?i)Arnbar.*", value="Armbar", regex=True) #armbar misspelling
methods = methods.replace(r"(?i)Ezekiel.*", value="Ezekiel Choke", regex=True) #ezekiel misspelling
methods = methods.replace(r"(?i)Ezequiel.*", value="Ezekiel Choke", regex=True) #ezekiel misspelling
methods = methods.replace(r"(?i)Guillotina.*", value="Guillotine", regex=True) #Guillotine misspelling
methods = methods.replace(r"(?i)Guilotine.*", value="Guillotine", regex=True) #Guillotine misspelling
methods = methods.replace(r"(?i)Adv.*", value="Referee Decision", regex=True) #adv/ref decision misspelling
methods = methods.replace(r"(?i)Pts: 0X0.*", value="Referee Decision", regex=True) #adv/ref decision misspelling
methods = methods.replace(r"(?i)Br.*a.*", value="Bread Cutter Choke", regex=True) #bread cutter misspelling
methods = methods.replace(r"(?i)Anaconda.*", value="Anaconda Choke", regex=True) #Anaconda misspelling
methods = methods.replace(r"(?i)Amassa Pao.*", value="Amassa Pao Choke", regex=True) #Amassa Pao misspelling
methods = methods.replace(r"(?i)7Yo.*", value="7-Yo-Choke", regex=True) #7-Yo-Choke misspelling
methods = methods.replace(r"(?i)Smother.*", value="Smother Choke", regex=True) #Smother misspelling
methods = methods.replace(r"(?i)Policeman.*", value="Policeman Lock", regex=True) #policeman lock misspelling
methods = methods.replace(r"(?i)Leg.*", value="Leg Lock", regex=True) #leg lock misspelling
methods = methods.replace(r"(?i)Choke F.*Omoplata", value="Choke From Omoplata", regex=True) #choke from omoplata misspelling
methods = methods.replace(r"(?i)Wrislock", value="Wristlock", regex=True) #wrist lock misspelling
methods = methods.replace(r"(?i)Verbal.*", value="Verbal Tap", regex=True) #verbal misspelling
methods = methods.replace(r"(?i)Darce.*", value="Darce Choke", regex=True) #darce choke misspelling
methods = methods.replace(r"---", value="N/A", regex=True) #---
methods = methods.fillna('N/A')

pd.set_option('display.max_rows', None)
df['method'] = methods
ndf = df.groupby('method')['method'].value_counts()
print(ndf)
```

```python
#weightDiv
weights = df['weightDiv']
weights = weights.str.upper()
weights = weights.str.strip('U')
weights = weights.replace(r'100KG+', value="O100KG")
weights = weights.replace(r'82KF', value="82KG")
weights = weights.replace(r'88.*', value="88KG", regex=True)
weights = weights.replace(r'62 KG', value="62KG")
weights = weights.replace(r'HWA', value="HWABS")
weights = weights.replace(r'LWA', value="LWABS")
weights = weights.replace(r'F|SF|3RD|4F|R1|//KG|SPF|---', value="N/A", regex=True)

print(weights.value_counts())
print(weights.nunique())

df['weightDiv'] = weights
```
